```python
import matplotlib.pyplot as plt
plt.figure(figsize=(6,4))
rows,cols = 2,3
fig,ax = plt.subplots(rows,cols,sharex='col',sharey='row')
for row in range(2):
    for col in range(3):
        ax[row,col].text(0.5,0.5,str((row,col)),ha='center')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0,2*np.pi,400)
y = np.sin(x**2) + np.cos(x)
derivative = 2*x*np.cos(x**2) - np.sin(x)

f,(ax1,ax2) = plt.subplots(1,2,sharey=True)
ax1.plot(x,y)
ax1.set_title('Sharing Y axis')
ax2.plot(x,derivative)
```

```python

fig,axes = plt.subplots(2, 2, subplot_kw=dict(polar=True))
axes[0,0].plot(x,y)
axes[0,1].plot(x,np.sin(x**2) + np.cos(x**2))
axes[1,0].plot(x,np.cos(x)*np.sin(x**2))
axes[1,1].plot(x,derivative,'g--')
```

```python
gaussian_numbers = np.random.normal(size=10000)
n,bins,patches = plt.hist(gaussian_numbers)
plt.title('GGG')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```
