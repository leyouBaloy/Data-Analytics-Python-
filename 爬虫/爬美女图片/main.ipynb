```python
#获取网页
import requests
headers = {
    'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.141 Safari/537.36'
}
response = requests.get('https://www.tupianzj.com/meinv/mm/qingchunmm/')
response.encoding='gb2312'#解决response.txt乱码问题
# print(response.request.headers)
print(response.text)
```

```python
#解析网页，得到美女图片地址
import re
html = response.text
urls = re.findall('<img src="(.*?)" alt=".*?" border="0" .><.a>',html)
print(urls)
```

```python
#保存到当前目录
import time
for url in urls:
    time.sleep(1)
    #图片的名字
    file_name = url.split('/')[-1]
    response = requests.get(url,headers=headers)
    with open(file_name,'wb') as f: #wb以二进制的形式写入
        f.write(response.content)
```
